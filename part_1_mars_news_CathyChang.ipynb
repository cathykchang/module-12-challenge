````python
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup as bs
from webdriver_manager.chrome import ChromeDriverManager
import requests
````

````python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
````

````python
# Visit the Mars NASA news site: https://redplanetscience.com
url = 'https://redplanetscience.com'
````

````python
response = requests.get(url)
````

### Step 2: Scrape the Website

Create a Beautiful Soup object and use it to extract text elements from the website.

````python
# Create a Beautiful Soup object
soup = bs(response.text, 'html.parser')
````

````python
# Print formatted version of the soup
print(soup.prettify())
````

````python
# Extract title elements
soup.title
````

````python
# Extract just the text of the title elemt
soup.title.text
````

````python
# Extract the body element
soup.body
````

````python
# Extract just the text of the body element
soup.body.text
````

````python
# Text of the body
soup.body.p.text
````

````python
# Extract the paragraph heading elements
soup.body.find_all('p')
````

````python
# Paragraph text
paragraphs = soup.find_all('p')
for paragraph in paragraphs:
    print(paragraph.text)
````
